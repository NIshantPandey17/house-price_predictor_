```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\IIT-Bhu Iternship\Python\Jupyter Files\areaprice071.csv")
x=df[["area"]]
y=df["price"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Area price Model")
plt.xlabel("value of area")
plt.ylabel("Price")
plt.scatter(x,y,color="orange")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
model.fit(x,y)
```

```python
model.predict([[5]])
```

```python
import pickle
```
